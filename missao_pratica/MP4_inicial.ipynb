Importação das Lib's

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

Valores dos itens do grafico

```python
AL = [(1,100),(2,0),(3,0),(4,150),(5,0)]
VE = [(1,0),(2,40),(3,50),(4,0),(5,50)]
TR = [(1,0),(2,0),(3,100),(4,300),(5,50)]
```

Função para 'maquiar' os plot

```python
def make_plot(plot_data,color,label):
  data_array = np.array(plot_data)
  transposed = data_array.T
  x,y = transposed 
  plt.plot(x,y, color=color, label=label, marker='o')
  
```

```python
make_plot(AL,"blue","Alimentação")
make_plot(VE,"red","Vestuario")
make_plot(TR,"green","Transporte")
```

Chamadas finais para exibir o primeiro grafico

```python

plt.xlim(1)
plt.title("Graficos de despesas")
plt.xlabel("DIA")
plt.ylabel("Despesas em R$")
plt.legend()
plt.show
```

desmembrar o dados do alimentos

```python
y = np.array([100,0,0,150,0])
x = np.array([1,2,3,4,5]).reshape((-1,1))
model = LinearRegression().fit(x,y)
y_pred = model.predict(x)
```

Chamadas finais para exibir o segundo grafico

```python
plt.scatter(x,y,c="red")
plt.plot(x,y,c="blue")
plt.plot(x,y_pred,c="black")
plt.legend()
plt.show()
```
